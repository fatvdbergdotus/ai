```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import os

# Set your Hugging Face API token (you can get it from https://huggingface.co/settings/tokens)
# It's recommended to set it as an environment variable for security.
os.environ["HF_AUTH_TOKEN"] = "PUT_YOUR_HUGGING_FACE_API_TOKEN_HERE"

# Load the pre-trained LLaMA model and tokenizer (replace with the correct model name)
# model_name = "meta-llama/Llama-3.1-8B-Instruct"  # Make sure this model exists on Hugging Face
model_name = "meta-llama/Llama-3.2-1B"

tokenizer = AutoTokenizer.from_pretrained(model_name, token=os.getenv("HF_AUTH_TOKEN"))
model = AutoModelForCausalLM.from_pretrained(model_name, token=os.getenv("HF_AUTH_TOKEN"))

tokenizer.pad_token = tokenizer.eos_token

# Define a simple prompt
prompt = "Once upon a time there was a "

# Tokenize the input text and generate the output
input_ids = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True).input_ids

# Generate output using the model
output = model.generate(input_ids, max_length=150, num_beams=4, temperature=0.7,pad_token_id=tokenizer.pad_token_id)

# Decode the generated tokens into text
output_text = tokenizer.decode(output[0], skip_special_tokens=True)

# Print the generated text
print(output_text)

```

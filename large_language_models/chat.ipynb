```python
import ollama
output = ollama.chat(model="llama3", messages=[
    {"role": "user", 
     "content": "What is the capital of USA and list all US states?"
     }])
print(output["message"]["content"])
```

```python
import ollama
output = ollama.chat(model="llama3", messages=[
    {"role": "user", 
     "content": "What is the capital of USA and list all US states?"
     }], stream=True)
for chunk in output:
    print(chunk["message"]["content"], end=" ", flush=True)
```

```python
from ollama import Client
ollama_client = Client(host="http://localhost:11434")
output = ollama_client.chat(model="llama3", messages=[
    {"role": "user", 
     "content": "What is the capital of USA?"
     }])
print(output["message"]["content"])
```

```python
from ollama import Client
ollama_client = Client(host="http://localhost:11434")
output = ollama_client.embed(model="llama3", input=["dog","puppy"])
print(output["embeddings"])
```
